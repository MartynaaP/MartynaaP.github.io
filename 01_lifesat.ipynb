```python
# prompt: Get data and visualize
# Download lifesat data into pandas dataframe called lifesat available at "https://github.com/marcinsawinski/UEP_KIE_ML_CLASS/raw/main/datasets/lifesat/lifesat.csv"
# Visualize with scatter plot "GDP per capita (USD)" on x axis , "Life satisfaction" of y axis

import pandas as pd
import matplotlib.pyplot as plt

# Download lifesat data into pandas dataframe called lifesat
lifesat = pd.read_csv("https://github.com/marcinsawinski/UEP_KIE_ML_CLASS/raw/main/datasets/lifesat/lifesat.csv")

# Visualize with scatter plot "GDP per capita (USD)" on x axis , "Life satisfaction" of y axis
plt.scatter(lifesat["GDP per capita (USD)"], lifesat["Life satisfaction"])
plt.xlabel("GDP per capita (USD)")
plt.ylabel("Life satisfaction")
plt.show()
```

# Task 1-1.2
_Train the model and predict_
- Assign new variable X with data from "GDP per capita (USD)" column converted to numpy array
- Assign new variable y with data from "Life satisfaction" column converted to numpy array
- Train linear regression Model for X and y and predict "Life satisfaction" when "GDP per capita (USD)" is equal 37655.2
- Train k-nearest neighbors Model with k=3 for X and y and predict "Life satisfaction" when "GDP per capita (USD)" is equal 37655.2
- Compare linear regression k-nearest neighbors results

_Type your code below_

```python
# prompt: Assign new variable X with data from "GDP per capita (USD)" column converted to numpy array
# Assign new variable y with data from "Life satisfaction" column converted to numpy array
# Train linear regression Model for X and y and predict "Life satisfaction" when "GDP per capita (USD)" is equal 37655.2
# Train k-nearest neighbors Model with k=3 for X and y and predict "Life satisfaction" when "GDP per c

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

# Assign new variable X with data from "GDP per capita (USD)" column converted to numpy array
X = np.array(lifesat["GDP per capita (USD)"]).reshape(-1, 1)
#reshape zmienia rozmiar tablicy (2,3) -> bedą dwa wiersze, 3 kolumny
#reshape(-1,1)-> zrób tyle wierszy ile potrzeba, ale 1 kolumnę
# Assign new variable y with data from "Life satisfaction" column converted to numpy array
y = np.array(lifesat["Life satisfaction"])

# Train linear regression Model for X and y and predict "Life satisfaction" when "GDP per capita (USD)" is equal 37655.2
linear_model = LinearRegression()
linear_model.fit(X, y)
linear_prediction = linear_model.predict([[37655.2]])
print("Linear Regression Prediction:", linear_prediction[0])

# Train k-nearest neighbors Model with k=3 for X and y and predict "Life satisfaction" when "GDP per capita (USD)" is equal 37655.2
knn_model = KNeighborsRegressor(n_neighbors=3)
knn_model.fit(X, y)
knn_prediction = knn_model.predict([[37655.2]])
print("K-Nearest Neighbors Prediction:", knn_prediction[0])

# Compare linear regression and k-nearest neighbors results
# (You can add more detailed comparison if needed, e.g., analyzing model performance metrics)
print("Comparison:")
print("Linear Regression Prediction:", linear_prediction[0])
print("K-Nearest Neighbors Prediction:", knn_prediction[0])
```
